```python
# remove outliers
upper_limit = df['Rent'].quantile(0.99)
df = df[df['Rent'] <= upper_limit].copy()
df
```


Problem: Rent was highly right-skewed due to extreme high values.
How detected:
Histogram showed long right tail
Max value (35 lakh) was unrealistic
Fix:
Removed values above 99th percentile
Applied log transformation
Result:
Skew reduced
Distribution closer to normal
Better for regression models
Lesson: Always check distribution before modeling.

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Rent'], kde=True)
plt.show()
```


Decision: Train model on log(Rent) instead of raw Rent.
Why: Target was right-skewed. Log transform reduced skew and stabilized variance.
Important: After prediction → apply exp() to convert back.
Common mistake to avoid: Forgetting inverse transform when evaluating real-world predictions.

```python
df['Rent_log'] = np.log(df['Rent'])
```

```python
df.columns
```


Why we check Size first: Size directly influences rental value. It’s continuous and likely linear.
EDA principle: Always test strongest logical feature first.

```python
df[["Size", "Rent_log"]].corr()
```


Observation: Correlation between Size and Rent_log = 0.51 (moderate).
Scatter shows: Upward trend but increasing variance.
Insight: Size influences rent, but location and quality likely also important.
Lesson: No single feature explains price fully.


When modeling house prices:
Hierarchy of influence (usually):
1️⃣ Location
2️⃣ Size
3️⃣ Property type / BHK
4️⃣ Furnishing
Lesson: Macro factors > micro features.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x= df['Size'], y = df['Rent_log'])
```

```python
df.groupby('City')['Rent'].median().sort_values(ascending=False)
```


Observation: City-wise median rent varies significantly. Mumbai median rent (~50k) is much higher than others.
Insight: Location (City) is a dominant feature.
Lesson: Categorical macro-features can be more powerful than numeric size.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='City', y="Rent_log", data=df)
plt.xticks(rotation=45)
plt.show()
```


Observation: City distributions are clearly separated.
Insight: City strongly affects baseline rent.
Modeling implication: City must be encoded properly (One-hot encoding).
Lesson: Categorical macro variables can create structural shifts in target distribution.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['BHK'], y=df["Rent_log"], data=df)
plt.show()
```

```python
df['BHK'].value_counts()
```

Problem: 5 & 6 BHK categories had extremely low sample size (7 each).
Risk: Model may overfit rare categories → high variance → poor generalization.
Fix: Combine rare categories (4+ BHK) to stabilize learning.
Lesson: Small sample categories increase model variance.

FINAL EDA SUMMARY – House Rent Prediction

1️⃣ Target Variable: Rent
Problem:
Highly right-skewed
Extreme values (up to 35 lakh)
Action:
Removed extreme outliers (top percentile)
Applied log transformation → Rent_log
Result:
Distribution became more normal
Variance stabilized
Better for Linear Regression



2️⃣ Size
Correlation with Rent_log ≈ 0.51 (moderate)
Clear upward trend
Bigger house → higher rent
Conclusion: ✔ Strong continuous predictor
✔ Keep as primary feature

3️⃣ BHK
1 → lowest rent
2 → higher
3 → higher
4+ → higher but very few samples
Problem:
5 & 6 BHK have very low counts (7 samples each)
Decision:
Combine rare categories (5+ into 4+)


4️⃣ Bathroom
Positive relationship with rent
But highly correlated with BHK
Correlation (Size vs BHK ≈ 0.70)
Insight: May cause multicollinearity in Linear Regression.
But: Tree models don’t care much.

5️⃣ City (Very Strong)
Clear separation:
Mumbai highest
Delhi moderate
Kolkata lowest
Conclusion: ✔ Major structural feature ✔ Must encode properly (One-hot)


7️⃣ Area Type
